**O `groupby` nos permite agregar os dados, o que facilita muito a visualização:**

```python
base.groupby("coluna1")["coluna2"].funcao_agregacao()





- O groupby vai agregar pela <font color="red">**coluna1**</font>
- A <font color="blue">**função de agregação**</font> será aplicada na <font color="red">**coluna2**</font>

**Postagens com pessoas engajam muito mais para essa marca, sendo 3 vezes maior de quando não tem pessoas**

Para conseguir analisar separadamente as tags, podemos dividir linhas com 2 tags em 2 linhas. Para isso primeiro vamos usar o split para separar em uma lista com as tags e depois vamos usar o explode para transformar as listas com 2 tags em 2 linhas diferentes. A primeira coisa que o split vai fazer é separar as informações. O split separa um texto em uma lista baseado em algum separador, mas, se não informamos nenhum argumento, vai separar por espaço.


Estrutura: texto.split() ou texto.split(‘delimitador’)

- Agora vamos utilizar esse comando para a
nossa coluna de Tags, transformando a
coluna Tags em uma lista de tags e separando as tags por “/”. base.Tags = base.Tags.str.split("/").

````python
#Vamos usar isso para a coluna "Tags"
#Transformando a coluna Tags em uma lista de tags
base.Tags = base.Tags.str.split("/")
base.head()
````

O explode vai separar uma coluna de um DataFrame em 1 linha para cada elemento da lista.


````python
#Criando os dicionários
dic = {
    "A": [[1,2], 3,[4,5,6], []],
    "B": [1,2,3,4],
}

# Tranformando esse dicionário em um DF
base_dic = pd.DataFrame(dic)
base_dic
````

````python
#Usando explode para separar a coluna A
base_dic = base_dic.explode('A')
````

Observe que na primeira o elemento A tem 2 elementos. O explode vai transformar a minha
primeira linha em 2 linhas, uma com o primeiro elemento e outra com o segundo. Estrutura: base_dic = base_dic.explode(‘Coluna’)

````python
base_dic
````

- Tudo que estiver em lista será separado em 1
linha por elemento da lista.
Se não tiver na lista, o elemento será mantido.
Listas vazias vão ter o valor de NaN.
Para as outras colunas, elas irão repetir os seus
valores.
Inclusive o índice também irá repetir.

Vamos aplicar o explode na nossa coluna
de Tag agora que ele já está separado:

````python
base = base.explode('Tags')
base.head()
````

- Feito isso agora podemos começar a trabalhar com as nossas Tags.

Agora vamos fazer o mesmo groupby que para o cálculo de média para pessoas. Repare que temos uma diferença. Como adicionamos linha, toda análise que estamos fazendo só pode ser feita para essa coluna nova de Tag. Por exemplo, se eu tinha uma foto com 5000 curtidas e 2 tags, agora ela vai aparecer 2 vezes:

Para melhorar a visualização podemos colocar em ordem decrescente os valores de curtidas. Para isso, vamos utilizar o comando sort_values:

sort_values(“Coluna",ascending=False)

````python
base.groupby("Tags")[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
````

Dessa forma, é possível notar que postagens de promoções são as que mais engajam. Além de promoções, datas comemorativas e trends também possuem um bom

E o que está sem tag? Como vamos tratar
isso?

O primeiro passo é identificar os valores sem Tag utilizando o filtro:

Estrutura: base[base.Tags.isnull()]

````python
base[base.Tags.isnull()]
````

Da mesma forma que fizemos para Carrossel, podemos fazer para as tags escrevendo
"Sem tag" e nesse caso iria aparecer no groupby.

````python
#Atribuindo o texto sem tag para as colunas onde a tag é NaN
base.loc[base.Tags.isnull(),'Tags'] = 'Sem Tags'
````

````python
base.groupby("Tags")[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
````

Agora vamos repetir a tabela de curtidas por Tag considerando valores sem tag.
Observe que a categoria Sem Tags aparece na quarta posição. Uma interpretação possível é que podem ser fotos muito específicas ou quando faço algo muito genérico e tem um engajamento bom.


````python
#Podemos voltar como NaN caso a gente queira somente ignorar esses valores conforme orientado.
base.loc[base.Tags == 'Sem Tags','Tags'] = np.nan
````

````python
# E voltamos com as colunas com valores nulos:
base[base.Tags.isnull()]
````

````python
#E essas linhas novamente param de ser consideradas na agregação.
base.groupby("Tags")[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
````

````python
#Trabalhando com a coluna "Pessoas" e a coluna "Tags":
base.groupby(["Pessoas", "Tags"])[["Curtidas","Comentários"]].mean()
````

````python
#Trabalhando por ordenamento de "Curtidas":
base.groupby(["Pessoas", "Tags"])[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
````

Podemos concluir que todas as tags com pessoas têm um valor bom. Comparando Novos Produtos com pessoas e sem pessoas, é possível observar que ele funciona muito melhor com pessoas.

````python
#Campanhas e Tags:
base.groupby(["Campanhas", "Tags"])[["Curtidas","Comentários"]].mean().sort_values("Curtidas",ascending=False)
````

Observe que algumas vezes, mesmo não tendo campanha, promoção funciona. Campanhas não é tão determinante assim em
